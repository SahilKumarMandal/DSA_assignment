**Q1.** Find the value of T(2) for the recurrence relation **T(n) = 3T(n-1) + 12*n**, given that **T(0)=5**

**Solution:**

Given: T(n) = 3T(n-1) + 12*n, T(0)=5
To find: T(2) = ?

put n = 1 in the given equation,

* T(1) = 3T(1 - 1) + 12 * 1

* => T(1) = 3T(0) + 12 = 3 * 5 + 12 
     
* => T(1) = 27    (given T(0) = 5)

therefore, T(1) = 27

Now, put n = 2 in the given equation,

* T(2) = 3T(2-1) + 12 * 2 
     
* => T(2) = 3T(1) + 24
     
* => T(2) = 3*27 + 24               (using, T(1) = 27)
     
* => T(2) = 105

Hence, **T(2) = 105**

**Q2.** Given a recurrence relation, solve it using the substitution method:

**a. T(n) = {1 ; n=1, T(n-1) + c ; n > 1}**

**Solution:** According to ques,

* T(n) = T(n-1) + c   ...(i)

* T(n-1) = T(n-2) + c     ...(replacing n by (n-1) in (i) )

* therefore, T(n) = T(n-2) + 2*c  ...(substituting the value of T(n-1) in (i))

from this we can visualise that, the general equation of T(n) will be:
* T(n) = T(n-k) + k*c   ...(ii)

According to given initial condition T(n)=1 if n = 1, so we can assume that
* if, n-k = 1 then, T(n-k) = 1 and k = n - 1        ...(iii)
* Therefore, T(n) = 1 + (n-1)*c = O(n)    ...( substituting (iii) in (ii) )

Hence, **Time complexity = O(n)**

* T(n) = 2^k*T(n/2^k) + k.n  ...(iv)

* T(n) = 1 if n = 1
* therefore,
* if n/2^k = 1 then, T(n/2^k) = 1 and k = log(n) of base 2      ...(v)

* The dominant term in this expression is n.(log(n) of base 2) because it grows more quikly that the linear term n as n becomes large.

thus, **Time complexity = O(n*log(n) of base 2)**

**Q3.** Given a recurrence relation, solve it using the recursive tree approach:

**a. T(n) = {1; n = 1, 2T(n-1) + 1 ; n>1}**

**solution:** According to ques,
we have,
* T(n) = 2T(n-1) + 1 ... (i)

replace n by n-1 in (i), we get
* T(n-1) = 2T(n-2) + 1  ...(ii)

put the value of T(n-1) from (ii) in (i), we get
* T(n) = 2^2T(n-2) + 2 + 1 ...(iii)

from (i) and (iii) we can assume that after k iterations
* T(n) = 2^kT(n-k) + 2^(k-1) + 2^(k-2) + ... + 2^(k-k)
* => T(n) = 2^kT(n-k) + 2^(k-1)(1 - 2^-k)/2^-1  ...{using formula, a^1 + a^2 + ... + a^n = a(r^n - 1)/(r-1), where r is common ratio}
* => T(n) = 2^kT(n-k) + 2^k - 1 ...(iv)

from the given initial condition, T(n) = 1 if n = 1. therefore
* if T(n-k) = 1 then, n-k = 1 and k = n-1   ...(v)

from v and iv
* T(n) = 2^(n-1)*1 + 2^(n-1) - 1
* => T(n) = 2^n - 1 ...(vi)

hence, from vi
* **Time complexity = O(2^n)**
